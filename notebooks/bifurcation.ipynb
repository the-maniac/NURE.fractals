Init libs

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Logistic function

```python
def logistic(r, x):
    return r * pow(x, 2) * (1 - x)
```

Logistic function graph

```python
iterations = 1000
last = 50
n = 10000
r = np.linspace(4, 27/4, n)
# r = np.linspace(5, 5.5, n)


x = np.linspace(0.6, 0.8, n)


fig, ax = plt.subplots(1, 1, figsize=(10, 10))
for i in range(iterations):
    x = logistic(r, x)
    if i >= (iterations - last):
        ax.plot(r, x, 'k')
# original bifurcation
ax.set_xlim(4, 27/4-0.5)

# confirmation r1
# ax.set_xlim(5.32, 5.34)
# ax.set_ylim(0.74, 0.76)

# visual detection of r2
# ax.set_xlim(5.76, 5.765)
# ax.set_ylim(0.825, 0.875)

ax.set_title("Bifurcation diagram")
# plt.xscale('log')
# plt.yscale('log')
# plt.tight_layout()
plt.show()
```

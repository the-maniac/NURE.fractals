```python
import matplotlib.pyplot as plt
import numpy as np


def P(x, p=0.7):
    if x >= 1: return 1
    if x <= 0: return 0
    if 0 < x <= 0.5:
        return p*P(2*x, p) + (1-p)*0
    else:
        return p*1 + (1-p)*P(2*x-1, p)

# Plot
N = 512
p = 0.7
f, ax = plt.subplots()
ax.plot(np.arange(0, 1, 1/N), [P(x/N, p) for x in range(0, N)], color="deepskyblue")
ax.set_xlabel('x')
ax.set_ylabel('P(x)')
ax.grid(True)
plt.show()
```
